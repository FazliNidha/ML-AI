```python
data["Intl_Plan"] = data['Intl_Plan'].map({"no":0,"yes":1})
```

```python
data["Vmail_Plan"] = data['Vmail_Plan'].map({"no":0,"yes":1})
```

```python
data["Churn"] = data['Churn'].map({"no":0,"yes":1})
```

```python
data.info()
```

```python
data.describe().T
```

```python
data2=data.drop(['Phone','State'],axis=1)
```

```python
x=data2.drop('Churn',axis=1)
y=data2["Churn"]
```

```python
X=pd.get_dummies(x)
```

```python
X.shape
```

```python
X
```

```python
data2
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.2,random_state=134,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train.loc[:,:] = sc.fit_transform(x_train)
x_test.loc[:,:] = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
rf = RandomForestRegressor()
lr = LinearRegression()
dt = DecisionTreeRegressor()
cv_results = cross_val_score(rf, x_train, y_train, cv= 5, scoring="r2")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params_rf = {'n_estimators': [ 200, 300,400],
             'max_depth': [8, 16, 32],
             'max_features': ['log2', 'sqrt']}
```
