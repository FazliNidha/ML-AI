```python
x=data[['Murder', 'Assault', 'UrbanPop', 'Rape']]
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4)
```

```python
km.fit(x)
```

```python
clusters=km.predict(x)
```

```python
pd.Series(clusters).value_counts()
```

```python
data["clusters"]=clusters
```

```python
data
```

```python
km.inertia_
```

```python
all_wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit(x)
  wcss=km.inertia_
  all_wcss.append(wcss)
```

```python
all_wcss
```

```python
plt.plot(range(1,11),all_wcss)
plt.show()
```

```python
km=KMeans(n_clusters=5)
km.fit(x)
clusters=km.predict(x)
```

```python
data["clusters"]=clusters
```

```python
data
```

```python
sns.scatterplot(x="UrbanPop",y="Rape",data=data,hue="clusters")
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
mergings=linkage(x,method="complete")
```

```python
plt.figure(figsize=(10,5))
dendrogram(mergings)
plt.show()
```

```python
labels=fcluster(mergings,85,criterion='distance')
```

```python
labels
```

```python
data['hc_clusters']=labels
```
